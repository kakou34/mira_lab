## Use the selected fixed image to register all

```python
# Define all necessary paths
base_path = Path('../').resolve()
train_set_path = base_path / 'data' / 'training-set'
params_path =  base_path / 'elastix' / 'parameter_maps'

train_imgs_dir = train_set_path / 'training-images'
train_labels_dir = train_set_path / 'training-labels'
train_masks_dir = train_set_path / 'training-mask'
output_path = train_set_path / 'registered_images'

# Define parameter maps to use
param_maps_to_use = ['Par0025rigid.txt']

# Define fixed imgage to use
fix_img_path = train_imgs_dir / '1001.nii.gz'
fix_name = fix_img_path.name.rstrip('.nii.gz')
for k, mov_img_path in enumerate(train_imgs_dir.iterdir()):        
    # Define inpupt/output paths
    mov_img_id = mov_img_path.name.rstrip('.nii.gz')
    mov_lab_path = train_labels_dir / f'{mov_img_id}_3C.nii.gz'
    mov_mask_path = train_masks_dir / f'{mov_img_id}_1C.nii.gz'
    for pm in param_maps_to_use:
        # Define output paths
        output_pm_path = output_path / pm.rstrip('.txt')
        
        out_img_path = output_pm_path / 'training-images'
        out_lab_path = output_pm_path / 'training-labels'
        out_mask_path = output_pm_path / 'training-mask'
        
        out_img_path.mkdir(exist_ok=True, parents=True)
        out_lab_path.mkdir(exist_ok=True, parents=True)
        out_mask_path.mkdir(exist_ok=True, parents=True)

        result_path = out_img_path / f'r_{mov_img_id}.nii.gz'
        res_lab_path = out_lab_path / f'r_{mov_img_id}_3C.nii.gz'
        res_mask_path = out_mask_path / f'r_{mov_img_id}_1C.nii.gz'

        # Don't register the fixed image
        if mov_img_path.name == fix_img_path.name:
            shutil.copy(mov_img_path, result_path)
            shutil.copy(mov_lab_path, res_lab_path)
            shutil.copy(mov_mask_path, res_mask_path)
            continue

        # Read and modify parameters file
        parameters_filename = params_path / pm
        field_value_pairs = [('ResultImageFormat', 'nii.gz'), ('FinalBSplineInterpolationOrder', '0.0')]
        utils.modify_field_parameter_map(field_value_pairs, parameters_filename)

        # Register
        transform_map_path = utils.elastix_wrapper(fix_img_path, mov_img_path, result_path, parameters_filename)

        # Correct transformation parameters file
        field_value_pairs = [('ResultImageFormat', 'nii.gz'), ('FinalBSplineInterpolationOrder', '0.0')]
        utils.modify_field_parameter_map(field_value_pairs, transform_map_path)

        # Transform labels
        utils.transformix_wrapper(mov_lab_path, res_lab_path, transform_map_path)

        # Transform brain_mask
        utils.transformix_wrapper(mov_mask_path, res_mask_path, transform_map_path)
```

## Generate Atlases

```python
# Define paths
base_path = Path('../').resolve()
base_path = base_path / 'data' / 'training-set'
pm_path = base_path / 'registered_images' / 'Par0025rigid'
img_path = pm_path / 'training-images'
lab_path = pm_path / 'training-labels'

# Reference values
ref_img_path = img_path / 'r_1001.nii.gz'
ref_img = sitk.ReadImage(str(ref_img_path))
ref_img_array = sitk.GetArrayFromImage(ref_img)

# Configs
n_classes = 4
n_volumes = len(list(lab_path.iterdir()))
ref_shape = ref_img_array.shape
labels_keys = {0: 'background', 1: 'csf', 2: 'wm', 3: 'gm'}

# Create acumulators
atlases = np.zeros(
    (n_classes, ref_shape[0], ref_shape[1], ref_shape[2])).astype('float32')
mean_vol = np.zeros_like(ref_img_array).astype('float32')

for i, img_filepath in tqdm(enumerate(img_path.iterdir()), total=len(list(img_path.iterdir()))):
    if img_filepath.name.endswith('txt'):
        continue
    
    # Get names
    img_name = img_filepath.name.rstrip('.nii.gz')
    labels_name = f'{img_name}_3C.nii.gz'
    
    # Acumulate mean image
    img = sitk.GetArrayFromImage(sitk.ReadImage(str(img_filepath)))
    img = utils.min_max_norm(img, 255).astype('uint8')
    mean_vol += img / n_volumes
    
    # Acumulate prob atlas image
    lab_img = sitk.GetArrayFromImage(
        sitk.ReadImage(str(lab_path/labels_name)))    
    for c in range(atlases.shape[0]):
        atlases[c, :, :, :] += np.where(lab_img == c, 1, 0)/n_volumes

# Get topological atlas by voting
topological_atlas = np.argmax(atlases, axis=0)

# Save results
output_path = base_path / "atlases" / "Par0025rigid"
output_path.mkdir(exist_ok=True, parents=True)
for c in range(atlases.shape[0]):
    utils.save_segementations(
        atlases[c, :, :, :], ref_img, output_path / f'p_atlas_{labels_keys[c]}.nii.gz')
    utils.save_segementations(
        np.where(topological_atlas == c, 255, 0),
        ref_img,
        output_path / f't_atlas_{labels_keys[c]}.nii.gz'
    )
utils.save_segementations(topological_atlas, ref_img, output_path / 't_atlas.nii.gz')
utils.save_segementations(mean_vol, ref_img, output_path / 'mean_volume.nii.gz')
```

## Get Tissue Models

```python
base_path = Path('../').resolve()
base_path = base_path / 'data' / 'training-set'
pm_path = base_path / 'registered_images' / 'Par0025rigid'
img_path = pm_path / 'training-images'
lab_path = pm_path / 'training-labels'
atlases_path = base_path / 'atlases' / 'Par0025rigid'
# Reference values
ref_img_path = img_path / 'r_1001.nii.gz'

t_atlas = sitk.GetArrayFromImage(sitk.ReadImage(str(atlases_path/'t_atlas.nii.gz')))
gt_labels = sitk.GetArrayFromImage(sitk.ReadImage(str(lab_path/'r_1001_3C.nii.gz')))

n_classes = 4
labels_keys = {0: 'background', 1: 'csf', 2: 'wm', 3: 'gm'}

t_histograms = np.zeros((4, 255))
gt_histograms = np.zeros((4, 255))

for i, img_filepath in tqdm(enumerate(img_path.iterdir()), total=len(list(img_path.iterdir()))):
    if img_filepath.name.endswith('txt'):
        continue
    mov_img_id = img_filepath.name.rstrip('.nii.gz')
    mov_lab_path = lab_path / f'{mov_img_id}_3C.nii.gz'
    
    img_array = sitk.GetArrayFromImage(sitk.ReadImage(str(img_filepath)))
    img_labels = sitk.GetArrayFromImage(sitk.ReadImage(str(mov_lab_path)))
    img_array = utils.min_max_norm(img_array, 255).astype('uint8')
    
    for c in range(n_classes):
        t_histograms[c, :] += np.histogram(img_array[t_atlas == c], bins=255, range=[0, 256])[0]
        gt_histograms[c, :] += np.histogram(img_array[img_labels == c], bins=255, range=[0, 256])[0]

t_histograms_density = t_histograms / np.sum(t_histograms, axis=1)[:, None]
gt_histograms_density = gt_histograms / np.sum(gt_histograms, axis=1)[:, None]

t_sum = np.sum(t_histograms_density, axis=0)[None, :]
gt_sum = np.sum(gt_histograms_density, axis=0)[None, :]
t_histograms_posterior = t_histograms_density / np.repeat(t_sum, 4, axis=0)
gt_histograms_posterior = gt_histograms_density / np.repeat(gt_sum, 4, axis=0)
```

```python
labels_keys = {0: 'Background', 1: 'CSF', 2: 'WM', 3: 'GM'}

plt.figure(figsize=(15,3))
plt.title('Probability density function of intensities for each class (based on majority voting atlas)')
for c in range(n_classes):
    plt.plot(t_histograms_density[c,:], label=labels_keys[c])
plt.ylim([0,0.05])
plt.ylabel('PDF')
plt.xlabel('Intensity')
sns.despine()
plt.legend()
plt.show()

plt.figure(figsize=(15,3))
plt.title('Probability density function of intensities for each class (based on reference labels)')
for c in range(n_classes):
    plt.plot(gt_histograms_density[c,:], label=labels_keys[c])
plt.ylim([0,0.05])
plt.ylabel('PDF')
plt.xlabel('Intensity')
sns.despine()
plt.legend()
plt.show()
```

```python
labels_keys = {0: 'Background', 1: 'CSF', 2: 'WM', 3: 'GM'}

plt.figure(figsize=(15,3))
plt.title('Posteriors (based on majority voting atlas)')
for c in range(n_classes):
    plt.plot(t_histograms_posterior[c,:], label=labels_keys[c])
plt.ylabel('Probability')
plt.xlabel('Intensity')
sns.despine()
plt.legend()
plt.show()

plt.figure(figsize=(15,3))
plt.title('Posteriors (based on reference labels)')
for c in range(n_classes):
    plt.plot(gt_histograms_posterior[c,:], label=labels_keys[c])
plt.ylabel('probability')
plt.xlabel('Intensity')
sns.despine()
plt.legend()
plt.show()
```
